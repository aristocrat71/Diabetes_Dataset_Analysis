## Diabetes Dataset

**Description**
- The data was collected from the Iraqi society, acquired from the laboratory of Medical City Hospital and (the Specializes Center for Endocrinology and Diabetes-Al-Kindy Teaching Hospital).
- Patients' files were taken and data extracted from them and entered in to the database to construct the diabetes dataset. The data consist of medical information, laboratory analysis.
- It includes 1000 patients and cover three classes (Diabetic, Non-Diabetic, and Predicted-
Diabetic).

*Diabetes is an opportune disease which has large wealth of data available and has with it huge complications. There is a need for a better and a more accurate approach in the diagnosis of the disease*
